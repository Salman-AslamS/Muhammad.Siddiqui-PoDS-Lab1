```python
#tb.fillna(data.mean)

# Step 1: Identify numeric columns
numeric_columns = tb.select_dtypes(include=['number']).columns

# Step 2: Calculate the mean for each numeric column
column_means = tb[numeric_columns].mean()

# Step 3: Replace NaN values in numeric columns with their respective means
tb[numeric_columns] = tb[numeric_columns].fillna(column_means)

# Resulting DataFrame
tb
tb.isna().sum()
```

```python
# Define a function to perform min-max scaling on a column
def min_max_scaling(column):
    min_val = column.min()
    max_val = column.max()
    return (column - min_val) / (max_val - min_val)

tb[numeric_columns] = tb[numeric_columns].apply(min_max_scaling)
tb
```
